### Analysis

Create a new dataframe called df2, and it should display the mean population of every year and group values by continent and sort them according to 2023 Population.

```python
df2 = df.groupby('continent')[['1970 population', '1980 population','1990 population', '2000 population','2010 population' , '2015 population', '2020 population', '2022 population', '2023 population' ]].mean().sort_values(by='2023 population', ascending=False)
df2
```

```python
# Get the Transpose of df2 and store in a new variable called df3
# Transpose: Columns are the index, and the index becomes the columns

df3 = df2.transpose()
df3
```

```python
# Plot the trends of the population in years
df3.plot(figsize=(15, 15))
```

```python
# Boxplots to Detect Outliers in the Population Growth
df3.boxplot(figsize=(15, 15))
```

# Filtering and Ordering

Selecting subsets of data that meet some conditions and Arranging the results in a certain cretaria


```python
# Generate Records according to specified countries

countries = ['Kenya', 'Uganda', 'Rwanda', 'Tanzania', 'Ethiopia']
condition = df['country'].isin(countries)

# Filter by the Condition
east_africa = df[condition]
east_africa.sort_values(by='2023 population', ascending=False)
```

```python
# Get All the Countries that Countains the Word "United"
condition = df['country'].str.contains('United')
df[condition]
```

```python
# Create a new dataframe df4, where the countries are the index
df4 = df.set_index('country')
df4
```

```python
# The Axis:
# 1. axis = 1: Refers to the columns
# 2. axis = 0: Refers to the rows

df.shape
```

```python
# Filter Columns based on the continent and cca3
df4.filter(items=['continent', 'cca3'], axis=1)
```

```python
df4.filter(items = ['Kenya', 'Uganda', 'Somalia'], axis=0)
```

```python
# Filtering based on Searching a term
df4.filter(like='United', axis=0)
```

```python
# The loc and iloc Filtering
# loc['value']: Filters based on the value
# iloc[index]: Based on the index
```

```python
# Loc information: United States
df4.loc['United States']
```

```python
df4.iloc[3]
```

```python
# Filter Based
# Countries Below Rank of 10, Sort Values By Continent(ascending) and Country(descending) in a Descending
df[df['rank'] <= 10].sort_values(by=['continent', 'country'], ascending=[True, False])
```
